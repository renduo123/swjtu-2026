```python
import math
import matplotlib.pyplot as plt

# 无附加质量计算
T_d1 = 0.20652
A1 = 0.192467
A4 = 0.180000

omega_d1 = 2 * math.pi / T_d1  # 30.41421 rad/s
delta1 = (1/3) * math.log(A1 / A4)  # 0.02225
zeta1 = delta1 / (2 * math.pi)  # 0.00354
omega_n1 = omega_d1 / math.sqrt(1 - zeta1**2)  # 30.41435 rad/s

# 有附加质量计算
T_d2 = 0.215
A1_add = 0.193252
A4_add = 0.184000
m_add = 0.006

omega_d2 = 2 * math.pi / T_d2  # 29.17771 rad/s
delta2 = (1/3) * math.log(A1_add / A4_add)  # 0.01610
zeta2 = delta2 / (2 * math.pi)  # 0.00256
omega_n2 = omega_d2 / math.sqrt(1 - zeta2**2)  # 29.17833 rad/s

# 联立方程求解
k_eq = (omega_n1**2 * omega_n2**2 * m_add) / (omega_n1**2 - omega_n2**2)
m_eq = k_eq / omega_n1**2

# 准备绘图数据
labels = ['无附加质量', '有附加质量']
damped_frequencies = [omega_d1, omega_d2]
undamped_frequencies = [omega_n1, omega_n2]

# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 绘制阻尼固有频率对比图
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(labels, damped_frequencies, color=['blue', 'orange'])
plt.title('阻尼固有频率对比')
plt.ylabel('频率 (rad/s)')

# 绘制无阻尼固有频率对比图
plt.subplot(1, 2, 2)
plt.bar(labels, undamped_frequencies, color=['blue', 'orange'])
plt.title('无阻尼固有频率对比')
plt.ylabel('频率 (rad/s)')

plt.tight_layout()
plt.show()
    
```

```python
import matplotlib.pyplot as plt


# 无附加质量的特征根
poles_without_mass = [-0.10777 + 30.41421j, -0.10777 - 30.41421j]

# 有附加质量的特征根
poles_with_mass = [-0.07479 + 29.17771j, -0.07479 - 29.17771j]

# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 绘制无附加质量的特征根
plt.scatter([pole.real for pole in poles_without_mass], [pole.imag for pole in poles_without_mass],
            color='blue', label='无附加质量', marker='x')

# 绘制有附加质量的特征根
plt.scatter([pole.real for pole in poles_with_mass], [pole.imag for pole in poles_with_mass],
            color='red', label='有附加质量', marker='o')

# 添加标题和标签
plt.title('有和无附加质量的特征根')
plt.xlabel('实部')
plt.ylabel('虚部')

# 显示网格
plt.grid(True)

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
