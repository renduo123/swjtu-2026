### 2. 生成一个4X5和一个5X4的矩阵，并计算它们的乘积。

```python
import numpy as np

# 生成一个4X5的矩阵
A= np.random.randint(1, 101, size=(4, 5))
print("4X5的矩阵：")
print(A)

# 生成一个5X4的矩阵
B= np.random.randint(1, 101, size=(5, 4))
print("5X4的矩阵：")
print(B)

# 计算两个矩阵的乘积
print("矩阵乘积的结果：")
print(A@B)
```

### 3. 函数 f 定义如下，其中x是任意实数: 𝑓(𝑥)=(sin⁡(𝑥+1)/𝑒^𝑥 )(𝑥+1)/2，生成一个数组a∈𝑅^5，计算b = f(a)。

```python
import numpy as np

# 生成一个长度为5的整数数组，元素范围是-100到100
a = np.random.randint(-100, 101, size=5)
print("数组a：", a)

# 定义函数f
def f(x):
    return (np.sin(x + 1) / np.exp(x)) * ((x + 1) / 2)

# 计算b = f(a)
b = f(a)
print("数组b = f(a)：", b)
```

### 4. 随机生成一个5行10列的矩阵，然后每行元素减去该行的平均值

```python
import numpy as np

# 随机生成一个5行10列的整数矩阵，元素范围是1到100
matrix = np.random.randint(1, 101, size=(5, 10))
print("原始矩阵：")
print(matrix)

# 计算每行的平均值，结果保留小数
row_means = matrix.mean(axis=1, keepdims=True)
print("每行的平均值：")
print(row_means)

# 每行元素减去该行的平均值
new_matrix = matrix - row_means
print("每行元素减去该行平均值后的矩阵：")
print(new_matrix)
```
