```python
def f(x):
    y = x ** 2 / 100
    return y

a = [1, 2, 3, 4, 5]
b = [1, 1, 1, 1, 1]
for i in range(len(a)):  # 遍历列表 a 的索引
    b[i] = f(a[i])      # 对 a 中每个元素调用 f(x)
print(b)
```
