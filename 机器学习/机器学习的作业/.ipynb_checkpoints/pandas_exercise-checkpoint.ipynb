## Step 6：数据中有缺失值吗？

```python
# 请填写相应代码
has_missing_values = titanic.isnull().any().any()
if has_missing_values:
    print("数据存在缺失值。")
else:
    print("数据不存在缺失值。")
```

## Step 7: 乘客的最大年龄和最小年龄是多少？

```python
# 请填写相应代码
min_age = titanic['Age'].min()
max_age = titanic['Age'].max()
print(f"最小年龄: {min_age}, 最大年龄: {max_age}")
```

## Step 8: 有多少人生还？

```python
# 请填写相应代码
titanic["Survived"].value_counts()
```

## Step 9: 男性和女性的生还比例分别是多少？

```python
# 请填写相应代码
titanic["Survived"].value_counts(normalize=1)
```

## Step 10: 按照船票价格降序排列

```python
# 请填写相应代码
titanic_sorted = titanic.sort_values(by='Fare', ascending=False)
print(titanic_sorted)
```

## Step 11: 绘制一个展示船票价格的直方图

```python
# 请填写相应代码
plt.rcParams['font.sans-serif'] = ['SimHei']# 设置 matplotlib 的字体为黑体，以支持中文显示

titanic = pd.read_csv('titanic.csv') 
plt.hist(titanic['Fare'], bins=15)
plt.xlabel('船票价格')  
plt.ylabel('频数')  
plt.title('船票价格分布直方图')  
plt.show() 
```

## Step 12: 绘制乘客年龄与船票价格的散点图

```python
# 请填写相应代码
sns.jointplot(x="Age",y="Fare",data=titanic)
```
