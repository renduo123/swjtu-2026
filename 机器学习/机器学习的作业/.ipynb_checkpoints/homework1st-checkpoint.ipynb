```python
# 任务1
famous_person = "Albert Einstein"#
message = "A person who never made a mistake never tried anything new."
print(f"{famous_person} once said, \"{message}\"")

# 任务2
list0 = [3, 9, 7, 11, 23, 5]
sum_for = 0
for num in list0:
    sum_for += num
print("For循环求和:", sum_for)

# 任务3
sum_while = 0
index = 0
while index < len(list0):
    sum_while += list0[index]
    index += 1
print("While循环求和:", sum_while)

# 任务4
divisible_by_3 = list(range(3, 31, 3))
print("能被3整除的数:", divisible_by_3)

# 任务5
def positive_square_sum(numbers):
    total = 0
    for num in numbers:
        if num > 0:
            total += num ** 2
    return total

# 任务6
test_list = [3, 5, -8, 2, 0, -4, 6]
result = positive_square_sum(test_list)
print("正数平方和:", result)

```
