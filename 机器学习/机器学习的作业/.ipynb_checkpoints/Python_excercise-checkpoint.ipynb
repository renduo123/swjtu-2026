### 2. 利用for循环计算以下列表元素的和

```python
list0 = [3, 9, 7, 11, 23, 5]
```

```python
sum=0
for a in list0:
    sum+=a
print(sum)
```

### 3. 利用while循环计算以下列表元素的和

```python
list0 = [3, 9, 7, 11, 23, 5]
```

```python
sum= 0
i= 0
while i < len(list0):
    sum+= list0[i]
    i+= 1
print(sum)
```

### 4. 创建一个列表，其中包含3~30 内能被3 整除的数字

```python
list1= []
number = 3
while number <= 30:
    if number % 3 == 0:
        # 如果能被 3 整除，就添加到列表中
        list1.append(number)
    number = number + 1
print(list1)
```

### 5. 编写一个函数，接收一个数字列表作为参数，计算其中所有正数的平方和并返回。

```python
def positive_square_sum(num_list):
    total = 0
    i= 0
    while i< len(num_list):
        # 获取当前索引对应的元素
        num = num_list[i]
        # 判断元素是否为正数
        if num > 0:
            # 如果是正数，计算其平方并累加到总和中
            total = total + num ** 2
        # 索引加 1，继续处理下一个元素
        i= i+ 1
    return total
```

### 6. 将列表[3, 5, -8, 2, 0, -4, 6]作为参数调用上面的函数并打印结果

```python
# 调用之前定义的函数
list_numbers = [3, 5, -8, 2, 0, -4, 6]
result = positive_square_sum(list_numbers)
print(result)
```
