## Matplotlib练习

### 1. 用蓝线绘制函数𝑓(𝑥) = 𝑥^2−2𝑥+3的图，并在坐标(1,2)位置处标上一个红色的点


```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-100,101)
y=x**2-2*x+ 3
plt.plot(x,y,"b",1,2,"ro")                                                       
plt.show()
```

### 2. 使用np.linspace()函数生成t∈[0, 2𝜋]。给定𝑥=16*sin⁡(𝑡)^3和𝑦=13*cos⁡(𝑡)−5*cos⁡(2𝑡)−2*cos⁡(3𝑡)−cos⁡(4𝑡)，画出x-y图，并给图标添加一个题目：Heart

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 t 值
t = np.linspace(0, 2*np.pi, 1000)  # 在 0 到 2π 之间生成 1000 个点

# 计算 x 和 y 的值
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# 绘制 x - y 图
plt.plot(x, y)

# 添加标题
plt.title('Heart')

# 显示图形
plt.show()
```
